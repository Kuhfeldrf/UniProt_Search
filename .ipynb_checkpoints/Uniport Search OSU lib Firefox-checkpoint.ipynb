```python
def download_website_as_xml(url, output_filename):
    # Step 1: Download the webpage content
    response = requests.get('url')
    response.raise_for_status()  # This will raise an exception if the request failed
    
    # Step 2: Use BeautifulSoup to parse the content
    soup = BeautifulSoup(response.content, 'lxml')
    
    # Step 3 and 4: Convert the parsed content to XML and write to a file
    with open(output_filename, 'wb') as file:
        file.write(soup.prettify("utf-8"))

# Test the function
#ebsite = 'https://search.library.oregonstate.edu/discovery/search?query=any,contains,10.1021%2Facs.jafc.0c00904&tab=Everything&search_scope=OSU_Everything_Profile&vid=01ALLIANCE_OSU:OSU&lang=en&offset=0'
```

```python
def search_osu_library_by_url(doi):
    base_url = "https://search.library.oregonstate.edu/discovery/search"
    params = {
        "query": f"any,contains,{doi}",
        "tab": "Everything",
        "search_scope": "OSU_Everything_Profile",
        "vid": "01ALLIANCE_OSU:OSU",
        "lang": "en",
        "offset": "0"
    }
    
    # Construct the URL
    search_url = base_url + "?" + "&".join(f"{key}={value}" for key, value in params.items())
    
    # Fetch the webpage (note: we won't execute the request here due to environment restrictions)
    # response = requests.get(search_url)
    # content = response.content
    
    return search_url  # For now, just returning the constructed URL

# Generate the search URL for the given DOI
search_url = search_osu_library_by_url("10.1021/acs.jafc.0c00904")
print(search_url)
```
